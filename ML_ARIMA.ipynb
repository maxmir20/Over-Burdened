```python
#Machine Learning using ARIMA to predict housing prices

import pandas as pd
from matplotlib import pyplot
import numpy as np

#importing the data file
houseprice = pd.read_csv('datathon_cleaning_1.csv', encoding='utf-8')
```

```python
houseprice= houseprice[houseprice['RoomType'] == 2]
houseprice
```

```python
# contains predictions for 2020, 2021, 2022
list_of_predictions = []

import pandas as pd
from matplotlib import pyplot
import numpy as np
from pandas.plotting import autocorrelation_plot
from pandas import datetime
from pandas import DataFrame
from statsmodels.tsa.arima_model import ARIMA
from matplotlib import pyplot
from statsmodels.tsa.arima_model import ARIMA
import pmdarima as pm


for i in range(34,82):
    data2 = houseprice.loc[i].T
    data2= data2[8:41]
    data2.index = pd.to_datetime(data2.index, format='%Y-%m')
    data2 = data2.astype(float)
    
    #using auto arima model 
    model = pm.auto_arima(data2, start_p=1, start_q=1,
                      test='adf',       
                      max_p=5, max_q=5, 
                      m=2,              
                      d=None,           
                      seasonal=False, 
                      start_P=0, 
                      D=0, 
                      trace=True,
                      error_action='ignore',  
                      suppress_warnings=True, 
                      stepwise=True)
    # Forecasting the prediction for the next 36 months
    n_periods = 36
    fc, confint = model.predict(n_periods=n_periods, return_conf_int=True)
    fc = fc.tolist()
    list_of_predictions.append(fc)
```

```python
#house_price_predictions dataframe has the predictions for the next few years
house_price_predictions = houseprice
```

```python
t2020 = []
t2021 = []
t2022 = []
for list1 in list_of_predictions:
    sum1 = 0
    sum2 = 0
    sum3 = 0
    for j in range(0,36):
        if(j <=11):
            sum1 += list1[j]
        elif( j >= 23):
            sum2 += list1[j]
        else:
            sum3 += list1[j]
    t2020.append(sum1/12)
    t2021.append(sum2/12)
    t2022.append(sum3/12)

house_price_predictions['2020']= t2020
house_price_predictions['2021']= t2021
house_price_predictions['2022']= t2022
```

```python
house_price_predictions
```

```python
# rent burden for an average household income in san francisco
income1 = 96000
householdinc = income1/12
```

```python
#rentburden in 2021
rentburden1 = house_price_predictions['2020']/householdinc
house_price_predictions['rentburden2020'] = rentburden1
```
